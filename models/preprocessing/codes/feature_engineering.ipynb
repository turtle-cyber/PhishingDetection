```python
# step0_requirements_and_imports.py

import re
import math
import json
import joblib
import tldextract
import numpy as np
import pandas as pd
from urllib.parse import urlparse, parse_qs
from collections import Counter
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

```

```python
# step1_load_and_initial_cleaning.py
# Load CSV (adjust path if needed)
INPUT_CSV = "C:\\Users\\HP\\Desktop\\PhishingDetection\\combined_features.csv"
df_raw = pd.read_csv(INPUT_CSV)

# Keep original URLs for reference
df_raw['original_url'] = df_raw.get('original_url')  # ensure column exists

# Drop rows with no URL (we can't extract features)
df = df_raw.dropna(subset=['original_url']).reset_index(drop=True).copy()
print("Rows after dropping missing URLs:", len(df))

# Quick look
display(df.head(3))
```

```python
# step2_helpers_feature_functions.py
# small utility functions used below

def shannon_entropy(s: str) -> float:
    if not s:
        return 0.0
    counts = Counter(s)
    length = len(s)
    ent = -sum((cnt/length) * math.log2(cnt/length) for cnt in counts.values())
    return ent

ipv4_re = re.compile(r'^\d{1,3}(\.\d{1,3}){3}$')
# simple IPv6 detect check
def is_ip_address(hostname: str) -> int:
    if not isinstance(hostname, str) or hostname == "":
        return 0
    if ipv4_re.match(hostname):
        return 1
    if ':' in hostname:  # naive IPv6 marker
        return 1
    return 0

def count_chars(s: str, predicate) -> int:
    if not isinstance(s, str):
        return 0
    return sum(1 for ch in s if predicate(ch))

def levenshtein(a: str, b: str) -> int:
    # Pure python dynamic programming (works fine for short strings)
    if a is None: a = ""
    if b is None: b = ""
    if len(a) < len(b):
        return levenshtein(b, a)
    if len(b) == 0:
        return len(a)
    prev = list(range(len(b)+1))
    for i, ca in enumerate(a, start=1):
        curr = [i]
        for j, cb in enumerate(b, start=1):
            ins = prev[j] + 1
            delete = curr[j-1] + 1
            sub = prev[j-1] + (0 if ca == cb else 1)
            curr.append(min(ins, delete, sub))
        prev = curr
    return prev[-1]

# tunnelling / serverless host suffixes to flag
TUNNEL_SUFFIXES = [
    "ngrok.io", "vercel.app", "github.io", "pages.dev", "herokuapp.com",
    "netlify.app", "now.sh", "zeit.co", "appspot.com"
]

def is_tunnelling(hostname: str) -> int:
    if not isinstance(hostname, str):
        return 0
    for s in TUNNEL_SUFFIXES:
        if hostname.endswith(s):
            return 1
    return 0
```

```python
# step3_parse_urls_and_create_derived_features.py
# We'll create derived columns and safe-guard if parsing fails
derived = []

# Optionally load a supplied CSE target domains list (for edit-distance)
# If you have a file with target CSE domains, load it; else provide a placeholder list.
# Example: cse_list = ["sbi.co.in", "icicibank.co.in", "bankofindia.in"]
CSE_LIST_PATH = None  # or "/path/to/cse_list.txt"
if CSE_LIST_PATH:
    with open(CSE_LIST_PATH) as f:
        cse_list = [ln.strip() for ln in f if ln.strip()]
else:
    # placeholder: replace with your real target CSE domains
    cse_list = ["sbi.co.in", "icicibank.com", "hdfcbank.com", "onlinesbi.sbi"]

for idx, row in df.iterrows():
    url = str(row.get('original_url') or "")
    parsed = urlparse(url if url else "")
    scheme = parsed.scheme.lower() if parsed.scheme else ""
    hostname = parsed.hostname or ""
    path = parsed.path or ""
    query = parsed.query or ""
    fragment = parsed.fragment or ""
    # tldextract to get registered domain, subdomain and suffix
    te = tldextract.extract(url)
    reg_domain = te.registered_domain or ""
    subdomain = te.subdomain or ""
    suffix = te.suffix or ""
    # lexical counts
    url_length = len(url)
    domain_length = len(hostname)
    path_length = len(path)
    query_length = len(query)
    num_dots = hostname.count('.') if hostname else 0
    num_hyphens = hostname.count('-')
    num_underscores = hostname.count('_')
    num_digits_domain = count_chars(hostname, str.isdigit)
    num_letters_domain = count_chars(hostname, str.isalpha)
    digit_ratio_domain = (num_digits_domain / domain_length) if domain_length>0 else 0.0
    letter_ratio_domain = (num_letters_domain / domain_length) if domain_length>0 else 0.0
    # special char counts on full url
    num_qm = url.count('?')
    num_eq = url.count('=')
    num_pct = url.count('%')
    num_at = url.count('@')
    num_dollars = url.count('$')
    num_exclam = url.count('!')
    # entropy
    domain_entropy = shannon_entropy(hostname)
    url_entropy = shannon_entropy(url)
    # https flag
    has_https = 1 if scheme == 'https' else 0
    # is ip
    ip_flag = is_ip_address(hostname)
    # tunnelling detection
    tunnel_flag = is_tunnelling(hostname)
    # min edit distance to CSE list (lower=closer)
    min_edit = min((levenshtein(hostname, cse) for cse in cse_list), default=999)
    # query param count
    try:
        qparams = parse_qs(query)
        num_query_params = len(qparams)
    except Exception:
        num_query_params = 0

    derived.append({
        "parsed_scheme": scheme,
        "hostname_parsed": hostname,
        "registered_domain_parsed": reg_domain,
        "subdomain_parsed": subdomain,
        "suffix_parsed": suffix,
        "url_length_derived": url_length,
        "domain_length_derived": domain_length,
        "path_length_derived": path_length,
        "query_length_derived": query_length,
        "num_dots_derived": num_dots,
        "num_hyphens_derived": num_hyphens,
        "num_underscores_derived": num_underscores,
        "num_digits_domain": num_digits_domain,
        "num_letters_domain": num_letters_domain,
        "digit_ratio_domain": digit_ratio_domain,
        "letter_ratio_domain": letter_ratio_domain,
        "num_qm": num_qm,
        "num_eq": num_eq,
        "num_pct": num_pct,
        "num_at": num_at,
        "num_dollars": num_dollars,
        "num_exclam": num_exclam,
        "domain_entropy": domain_entropy,
        "url_entropy": url_entropy,
        "has_https": has_https,
        "is_ip_hostname": ip_flag,
        "is_tunnelling": tunnel_flag,
        "min_edit_to_cse": min_edit,
        "num_query_params": num_query_params,
    })

df_derived = pd.DataFrame(derived, index=df.index)
# join to original df (do not overwrite original_url or existing important columns)
df = pd.concat([df.reset_index(drop=True), df_derived.reset_index(drop=True)], axis=1)
print("Derived features added. New shape:", df.shape)
display(df.loc[:, ['original_url','hostname','registered_domain_parsed','url_length_derived','domain_entropy','min_edit_to_cse']].head())
```

```python
# step4_merge_and_clean_existing_numeric_cols.py
# Identify numeric columns from crawler output and keep them
# (We will keep all numeric crawler columns + our derived columns; drop big text/ID columns)
cols_to_exclude = ['original_url', 'url_norm', 'full_html', 'rendered_html', 'screenshot_path']  # adapt to your actual columns
cols_to_exclude = [c for c in cols_to_exclude if c in df.columns]

# Build working feature list: numeric + newly derived
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
# ensure we include our derived numeric columns (they are numeric)
feature_cols = sorted(list(set(numeric_cols)))  # remove duplicates

# sanity: drop any index-like columns if present (e.g., unnamed)
feature_cols = [c for c in feature_cols if not str(c).lower().startswith('unnamed')]

print("Number of numeric features to keep:", len(feature_cols))
print("Some features:", feature_cols[:40])

# Fill missing numeric with 0 (safer for counts/flags)
df[feature_cols] = df[feature_cols].fillna(0)
```

```python
# step5_build_sklearn_preprocessing_pipeline.py
# We will define categorical columns (low-cardinality) present in parsed fields
# and numeric columns as the rest from feature_cols.

# Determine categorical candidate columns (small cardinality object columns +
# parsed fields we created)
cat_candidates = []
for c in ['parsed_scheme','hostname_parsed','registered_domain_parsed','subdomain_parsed','suffix_parsed']:
    if c in df.columns:
        cat_candidates.append(c)

# Convert those to string type (safe)
for c in cat_candidates:
    df[c] = df[c].fillna("unknown").astype(str)

# Decide categorical encoding strategy: low-cardinality -> OneHot, high-cardinality -> Ordinal
CAT_OHE_MAX_UNIQUE = 30
ohe_cols = []
ord_cols = []
for c in cat_candidates:
    nuni = df[c].nunique()
    if nuni <= CAT_OHE_MAX_UNIQUE:
        ohe_cols.append(c)
    else:
        ord_cols.append(c)

# Final numeric columns (exclude categorical and excluded)
numeric_pipeline_cols = [c for c in feature_cols if c not in ohe_cols + ord_cols]

print("one-hot cols:", ohe_cols)
print("ordinal cols:", ord_cols)
print("numeric cols count:", len(numeric_pipeline_cols))

# Build transformers
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

ohe_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='unknown')),
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse=False))
])

ord_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='unknown')),
    ('ord', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_pipeline_cols),
    ('ohe', ohe_transformer, ohe_cols) if ohe_cols else ('ohe', 'drop', []),
    ('ord', ord_transformer, ord_cols) if ord_cols else ('ord', 'drop', [])
], remainder='drop', sparse_threshold=0)

# Fit preprocessor on the whole dataset (or fit on training split when labels are available)
preprocessor.fit(df)

# Transform into numpy matrix
X_trans = preprocessor.transform(df)

# Try to get feature names (sklearn >= 1.0 supports get_feature_names_out)
try:
    feature_names = preprocessor.get_feature_names_out()
    X_df = pd.DataFrame(X_trans, columns=feature_names, index=df.index)
except Exception:
    # fallback: use numeric names + OHE generic names
    X_df = pd.DataFrame(X_trans, index=df.index)
    feature_names = X_df.columns.tolist()

print("Processed data shape:", X_df.shape)
display(X_df.head())
```

```python
# step6_save_preprocessor_and_processed_features.py
# Save preprocessor for later use in training/prediction
joblib.dump(preprocessor, "/mnt/data/preprocessor_ps02_stage1.joblib")
print("Saved preprocessor -> /mnt/data/preprocessor_ps02_stage1.joblib")

# Save processed features (you may also save indexes to map back to original rows)
X_df.to_csv("/mnt/data/processed_features_stage.csv", index=False)
print("Saved processed features -> /mnt/data/processed_features_stage.csv")

# Save mapping of original_url -> processed row (helps generate evidence linkage later)
df[['original_url']].reset_index().to_csv("/mnt/data/feature_row_index_map.csv", index=False)
print("Saved mapping file -> /mnt/data/feature_row_index_map.csv")
```

```python
# step7_optional_split_and_save_for_training.py
# If you already have labels (e.g., a column 'label' with values 'phishing'/'benign'),
# create train/test splits. Otherwise skip this step and add labels later.

LABEL_COL = 'label'  # change to your label column name if available
if LABEL_COL in df.columns:
    y = df[LABEL_COL].values
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X_df, y, test_size=0.2, stratify=y, random_state=42)
    X_train.to_csv("/mnt/data/X_train.csv", index=False)
    X_test.to_csv("/mnt/data/X_test.csv", index=False)
    pd.Series(y_train, name='label').to_csv("/mnt/data/y_train.csv", index=False)
    pd.Series(y_test, name='label').to_csv("/mnt/data/y_test.csv", index=False)
    print("Saved train/test splits to /mnt/data")
else:
    print("No '{}' column found. Add labels and re-run splitting code.".format(LABEL_COL))
```

### MAIN


```python
# preprocessing_ps02_stage1.py
# Purpose: Clean & preprocess combined_features.csv for PS-02 Stage-1 tabular models
# Output: X_processed (ndarray), y (Series), fitted ColumnTransformer (preprocessor)
#         plus simple prints for shape sanity checks and feature accounting.

import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import joblib
from pathlib import Path

CSV_PATH = "/mnt/data/combined_features.csv"
SAVE_DIR = Path("/mnt/data")  # change if you want to save elsewhere
SAVE_DIR.mkdir(parents=True, exist_ok=True)

# ----------------------------
# 1) Load
# ----------------------------
df = pd.read_csv(CSV_PATH)

# ----------------------------
# 2) Drop fully-empty columns
# ----------------------------
empty_cols = [c for c in df.columns if df[c].isna().all()]
df_clean = df.drop(columns=empty_cols)

# ----------------------------
# 3) Drop meta / non-feature identifiers (kept outside model)
# ----------------------------
meta_cols_to_drop = [
    "original_url", "url_norm", "hostname", "registered_domain",
    "domain", "subdomain", "screenshot_error", "dom_drift_score"
]
meta_cols_to_drop = [c for c in meta_cols_to_drop if c in df_clean.columns]
meta_df = df_clean[meta_cols_to_drop] if meta_cols_to_drop else pd.DataFrame()
df_clean = df_clean.drop(columns=meta_cols_to_drop, errors="ignore")

# ----------------------------
# 4) Convert clearly numeric-like object columns where possible
#    (we attempt a strict conversion; if it fails, column stays categorical)
# ----------------------------
obj_cols = df_clean.select_dtypes(include="object").columns.tolist()

# Map common boolean strings to 0/1 first (helps conversion)
bool_map = {"true": 1, "false": 0, "yes": 1, "no": 0, True: 1, False: 0}
df_clean = df_clean.replace(bool_map)

converted_to_numeric = []
still_categorical = []
for c in obj_cols:
    try:
        df_clean[c] = pd.to_numeric(df_clean[c], errors="raise")
        converted_to_numeric.append(c)
    except Exception:
        still_categorical.append(c)

# ----------------------------
# 5) Separate target and features
# ----------------------------
if "label" not in df_clean.columns:
    raise ValueError("Expected a 'label' column with values {'benign','phishing'}")

# map labels to 0/1 (benign=0, phishing=1). 'suspected' will be derived via thresholds at inference.
y = df_clean["label"].map({"benign": 0, "phishing": 1}).astype(int)
X = df_clean.drop(columns=["label"])

# ----------------------------
# 6) Numeric imputation (median) before scaling
# ----------------------------
# After step 4, any columns of dtype number are numeric; the rest are categorical
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in X.columns if c not in num_cols]

# We’ll impute numerics here (median). Categoricals will be imputed by their pipeline later.
num_imputer = SimpleImputer(strategy="median")
X[num_cols] = num_imputer.fit_transform(X[num_cols])

# ----------------------------
# 7) Build preprocessing pipeline
#    - Numeric: StandardScaler
#    - Categorical: OneHotEncoder(handle_unknown='ignore', cap high cardinality with max_categories)
# ----------------------------
numeric_transformer = Pipeline(steps=[
    ("scaler", StandardScaler())
])

# NOTE: using 'sparse=False' for wider sklearn compat; do NOT use 'sparse_output' here.
categorical_transformer = Pipeline(steps=[
    ("encoder", OneHotEncoder(handle_unknown="ignore", sparse=False, max_categories=20))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_cols),
        ("cat", categorical_transformer, cat_cols),
    ]
)

# ----------------------------
# 8) Fit & transform
# ----------------------------
X_processed = preprocessor.fit_transform(X)

# ----------------------------
# 9) Save artifacts (optional but recommended)
# ----------------------------
joblib.dump(preprocessor, SAVE_DIR / "preprocessor.pkl")
joblib.dump(num_imputer, SAVE_DIR / "num_imputer.pkl")
X_path = SAVE_DIR / "X_processed.npy"
y_path = SAVE_DIR / "y_labels.npy"
np.save(X_path, X_processed)
np.save(y_path, y.values)

# ----------------------------
# 10) Report quick summary
# ----------------------------
print("\n=== CLEANING SUMMARY ===")
print(f"Original shape:           {df.shape}")
print(f"After dropping empty:     {df.shape[0]} x {df.shape[1] - len(empty_cols)} (dropped {len(empty_cols)} all-NA cols)")
print(f"Meta cols dropped:        {meta_cols_to_drop}")
print(f"Converted to numeric:     {len(converted_to_numeric)} columns")
print(f"Remaining categoricals:   {len(still_categorical)} columns")

print("\n=== PREPROCESS SUMMARY ===")
print(f"Numeric feature count:    {len(num_cols)}")
print(f"Categorical feature count:{len(cat_cols)}")
print(f"Encoded feature count:    {len(preprocessor.get_feature_names_out())}")
print(f"X_processed shape:        {X_processed.shape}")  # (n_samples, n_features)
print(f"y distribution:           {y.value_counts().to_dict()}")

print("\nSaved:")
print(f"  - preprocessor.pkl       -> {SAVE_DIR / 'preprocessor.pkl'}")
print(f"  - num_imputer.pkl        -> {SAVE_DIR / 'num_imputer.pkl'}")
print(f"  - X_processed.npy        -> {X_path}")
print(f"  - y_labels.npy           -> {y_path}")
```
